**Entrada:** Varilla de largo $n$ y un arreglo de precios de cada medida $i..n$

**Salida:** Retorna el máximo $rₙ$ que se puede obtener cortando la varilla y vendiendo sus partes.

# 1. Descripción del problema

El método de programación dinámica al igual que el paradigma de divide y vencerás, se centra principalmente en dividir un problema base en sub-problemas e ir resolviendo cada una por separado. Sin embargo, la peculiaridad que posee la programación dinámica es que no necesita resolver un sub-problema que ya fue resuelto con anterioridad, a diferencia de algoritmos que hacen uso del paradigma de divide y vencerás que utilizan una implementación recursiva para resolver todos los sub-problemas, independientemente si hay repetidos.

Una situación en particular en el que se usa la programación dinámica es el problema del corte de varillas. En el que consiste en obtener la máxima ganancia posible al dividir la varilla y vendiendo sus partes.

# 2. Descripción del algoritmo
Supongamos que tenemos una varilla. Podemos dividir la varilla en distintos largos, en el que cada largo posee un determinado precio. ¿Cómo puedo maximizar la ganacia?

Podemos dividir la varilla iniciar en $2^{n-1}$ maneras.  Por ejemplo una varrila de largo 4 podemos dividirla en 8 maneras distintas, con su precio correspondiente arriba.

image.png


Entonces para resolver este problema podemos resolver problemas más pequeños  del mismo. Una vez que realizamos el primer corte o la primera division de la varilla consideramos las dos piezas en dos intancias independientes del problema original. Por lo que la solución optima para el problema incorpora las soluciones optimas de ambas piezas, obteniendo el valor máximo.

Básicamente podemos decir que el problema de la varilla presenta una subestructura óptima, es decir, las soluciones optimas de un problema incorporan las soluciones optimas de subproblemas relacionados, que podemos resolver de manera independiente.

Con lo anteriormente dicho, podemos resolver el problema de manera recursiva.

Analizaremos dos formas de resolver el problema anteriormente dicho, una forma es implementando una función recursiva $CutRod$ y otra iterativa $Bottom-up$.

------------------------------
$CutRod:$

Si realizamos un análisis para obtener los distintos cortes y tener su máximo valor, conseguiremos la siguiente ecuación:

$rₙ = max(pᵢ+rₙ-ᵢ$)
en el que pᵢ representa el precio y rₙ-ᵢ representa al valor máximo obtenido al cortar en dos la varilla en piezas de tamaño $i$ y $n-i$, siendo $i = 1,2,...n-1$

image.png

Primero la función recibe un arreglo con los precios de cada largo de varilla y el largo del mismo. Establecemos un caso base de que si $n == 0$ no hay ganancia posible ya que el largo de la vara es 0. Entonces declaramos la variable **valorMax** y se inicializa con un -1 para la primera comparacion que se realizará. Entonces recorremos de 1 a $n$ obteniendo el **valorMax** aplicando la ecuación de arriba, en el que rₙ-ᵢ será nuestra misma función $cutRod$, ya que así obtendemos la maxima ganancia de ese corte respectivo. Finalmente retornamos **valorMax**. 

-----------------------------

$bottom-up$

Esta función es similar a la anterior pero en esta declararemos un arreglo en el que guardaremos los resultados obtenidos de cada sub-problema, con esto nos ahorraremos utilizar recursivamente la misma función y un ciclo for. Para hacer las comparaciones declararemos un valor $q$, que corresponde a nuestro máximo valor del sub-problema. Luego comparamos nuestro $q$ actual con el precio de la porción de varilla sumado con los resultados obtenidos de cada sub-problema, obteniendo el mayor valor.



```python
import numpy as np
def precio():
  precios = np.array([1,5,8,9])
  return precios
  # precios = np.array([1,5,8,9,10,17,17,20,24,30])
```

# 3. Código cutRod

```python
import sys

def cutRod(p,n,verbose):
  global comp
  comp += 1
  if n == 0:
    return 0
  valorMax = -1
  for i in range(1,n+1):
    valorMax = max(valorMax,p[i-1] + cutRod(p,n-i,verbose))
    if verbose == True:
      print(valorMax)
  return valorMax

precios = precio()
comp = 0
valor = cutRod(precios,len(precios),verbose = False)
print("valor máximo:",valor)
print("número recursiones:",comp - 1)
```

#3.1 Código BottomUpCutRod

```python
def bottomUpCutRod(precio,largo,num_subProblemas,verbose):
  #Crear arreglo con máximos
  resultados_subProblemas = []

  resultados_subProblemas.append(0) # r_0 = 0

  #Resolver cada sub-Problema de 1...largo
  for j in range(1,largo+1):
    if verbose == True:
      print("j:",j)
      print("__________")
      print()
    valorMax = -1
    for i in range(1,j+1):
      if verbose == True:
        print()
        print("i:",i)
      aux = valorMax
      # Ecuación para obtener el valorMax
      valorMax = max(valorMax,precio[i-1] + resultados_subProblemas[j-i]) 
      if verbose == True:
        k = precio[i-1]+resultados_subProblemas[j-i]
        print("precio","[",i-1,"]","+ resultados_subProblemas","[",j-i,"]")
        print(precio[i-1],"+",resultados_subProblemas[j-i],"=",k)
        print("Obtener mayor: ",str(aux),"|",str(k))
        print("valorMax:",valorMax)
      num_subProblemas += 1
    #Agregar el valorMax al arreglo de máximos
    resultados_subProblemas.append(valorMax)
    if verbose == True:
      print("resultados subProblemas:",resultados_subProblemas)
      print("-------------------------")
      print()
    # if verbose == True:
        # print("Resultados:",resultados_subProblemas)

  return resultados_subProblemas[largo],num_subProblemas

precios = precio()
num_problemas = 0
valor,num_problemas = bottomUpCutRod(precios,len(precios),num_problemas,verbose = False)
print("valor máximo:",valor)
print("Cantidad de subProblemas:",num_problemas)
```

# 4. Correctitud


**Inducción matemática**

Para probar la correctitud, utilizaremos el metodo de inducción matemática, básicamente consisten en que si queremos probar alguna declaración $P(n)$ debemos probar un par de condiciones.

1. Probar un caso base para $R(n)$ como $R(1)$
2. Probar que si $R(m)$ es verdad para m < n, entonces R(n) también lo es.

Tomaremos que $P$ es el arreglo de precios y $r$ el arreglo de máximos

Para el caso base sabemos que si n = 0 no hay subproblemas a resolver y el valor que se devuelve en todos los casos es 0

Cuando se inicia un largo para representar la varilla $i$, la función opera con normalidad separando el problema general en $m - 1$ subproblemas. En estos subproblemas se debe verificar que opción le da un mayor valor a la varilla, el precio máximo establecido, o un precio ajustado por cortes. En ambos casos, para la comparación de datos, resulta necesario encontrar la solución de los subproblemas para los distintos cortes, sin embargo, si se asume la funcióin como correcta, la solución a este subproblema se encontraría en el arreglo $r$. Pueden darso dos casos para la comparación descrita anteriormente:

1. En el caso de que $P[m]$ sea el mayor, significa que ningún valor que utilice cortes resulta en un monto mayor del valor de $P[m]$, por lo que este valor es correcto y se anexa al arreglo de soluciones.

2. Si $P[i] + r[m-i]$ es escogido, es porque su valor resultó más grande que el de $P[m]$, por lo que es posible considerar uno o más cortes para maximizar el precio. Este valor es correcto y se anexa al arreglo $r$.

Cabe destacar que el algoritmo siempre tendrá la solución de $r[i-1..0]$ debido a que este parte resolviendo los subproblemas desde el que tiene menor tamaño, iterando para guardar todos los resultados en un arreglo, al que accederán los futuros subproblemas.

A partir de lo anterior, es posible concluir que el algoritmo llegará a un caso basi siempre que lo necesite, y en otro caso utilizará los datos ya disponibles para hallar una respuesta correcta.

# 5. Tiempo de ejecución

La cantidad de sub-problemas a resolver se obtiene de la definición mostrada al inicio (retorno máximo). Tal y como indica aquella fórmula, es necesario obtener el retorno máximo de $r$<sub>$n-i$</sub>, siendo $i$ evaluado entre $[1..n]$. 

- $c_r = [1..n] = n$
  
  $c_r =$ cantidad de retornos a calcular. 

Sin embargo, no basta sólo con obtener la cantidad de retornos que se deben calcular. Se debe tomar en cuenta los sub-problemas que resuelven los propios sub-problemas, es decir, la cantidad de iteraciones que realiza el **acercamiento Bottom-Up** en función de llegar del problema más pequeño, al más grande.

Esto se puede describir con la siguiente sumatoria:

- $c_s = \displaystyle\sum_{i = 1}^{n} $$r$<sub>$n-i$</sub>

 con $c_s =$ cantidad de sub-problemas y $r[0] = 0$.

Para desarrollar esta sumatoria, hay que dejar en claro la operación para calcular un retorno depende de $n$, y que si este retorno es un caso base, entonces sería una operación constante, por lo que considerando la expresión que permite definir la cantidad de retornos a calcular, se obtiene que:

- Si $c_s = $ $r$<sub>$n$</sub> entonces, $c_s = n$.

- Por lo tanto: $c_s = \displaystyle\sum_{i = 1}^{n} $$r$<sub>$n-i$</sub> 
 
 $ c_s = \displaystyle\sum_{i = 1}^{n} r_n + $$r$<sub>$n-1$</sub> $+$ $r$<sub>$n-2$</sub> $+$ $... r_0$ , con $r[0] = 0$

- Es posible expresar una nueva sumatoria a partir de los retornos (sin contar el $r[0]$ pues no afecta en la suma final):

 $ c_s = \displaystyle\sum_{i = 1}^{n} $$c$<sub>$ri$</sub> , con $c$<sub>$ri$ siendo la cantidad de retornos en $r_i$.

- A su vez, $c$<sub>$ri$</sub> puede expresarse según su resultado, el cual depende del largo de aquel sub-problema. Entonces, la sumatoria quedaría:

 $ c_s = \displaystyle\sum_{i = 1}^{n} $$c$<sub>$ri$</sub> , con $c$<sub>$ri$</sub> $= i$

 $ c_s = n + (n-1) + (n-2) + ... + 1$

 $c_s = n + \frac{n(n-1)}{2}$

 $c_s = \frac{n^2}{2} + \frac{n}{2}$

Por lo que en conclusión, la cantidad de sub-problemas a resolver se puede expresar en función al largo del arreglo (tamaño del problema) como: $\frac{n^2}{2} + \frac{n}{2}$.

El algoritmo posee dos ciclos for para resolver cada subproblema, obteniendo una complejidad temporal de O(n). Entonces el tiempo de ejecución de todo el algoritmo es de O(n^2), al tener que resolver los subproblemas de cada problema.

Cabe destacar que este algoritmo posee una complejidad temporar de O(n), ya que para poder obtener el máximo valor de manera iterativa se necesita crear un arreglo que guarde los resultados obtenidos de cada subproblema.

## 6.1 Tiempo de ejecución de $bottomUpCutRod$ y $cutRod$

Para realizar este experimento usaremos la función **cutrod_instance_generator**, que nos permitirá generar el precio correspondiente de cada varilla de tamaño $i$. Se trabajará con un tamaño del problema de 1 a 24

```python
import random

def cutrod_instance_generator(N):
  A = []
  prev = 0
  for i in range(N):
    r=random.randint(0,10)
    A.append(prev+r)
    prev+=r
  return A

cutrod_instance_generator(10)
```

```python
from matplotlib import cm
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
import random

x=[]; y=[]
ix=[]; by=[]


for n in range(1,25):
  arregloDePrecios = cutrod_instance_generator(n)
  verbose = False
  num_problemas = 0
  z = repeat(setup="from __main__ import cutRod", stmt=f"cutRod({arregloDePrecios},{n},{verbose})", repeat=1, number=10)

  x.append(n)
  y.append(z)

  q = repeat(setup="from __main__ import bottomUpCutRod", stmt=f"bottomUpCutRod({arregloDePrecios},{n},{num_problemas},{verbose})", repeat=1, number=10)
  ix.append(n)
  by.append(q)


plt.plot(x,y)
plt.plot(ix,by)

plt.legend(["CutRod","bottomUp"])

plt.xlabel('tamaño del arreglo')
plt.ylabel('Tiempo en Milisegundos')
plt.show()
```

Como podemos observar, el algoritmo **bottomUpCutRod** es claramente mucho más eficiente que el algoritmo **cutRod**. Esto tiene sentido al observar ambas complejidades temporales. Recordemos que **bottomUpCutRod** posee un tiempo de ejecución de $O(n^2)$ y **cutRod** un tiempo de $O(2^n)$. Con este experimento se comprueba que la programación dinamica es extremadamente útil en ciertos casos, obteniendo un tiempo de ejecución decente. Por ejemplo el tiempo que tomó crear este gráfico en custion fue de casi 4 minutos debido a las grandes cantidades de llamadas recursivas que realiza la función **cutRod**.
