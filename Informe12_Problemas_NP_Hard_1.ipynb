A continuación, se responderán las preguntas planteadas en el punto uno de la guía para este informe, las cuales son las siguientes:

- ¿ Qué quiere decir que un problema sea **NP-Hard**?

##Resolución

###¿Qué quiere decir que un problema sea **NP-Hard**?

En base a lo estudiado, se puede decir que un problema es **NP-Hard** cuando se asume que la conjetura $P \neq NP$ es verdadera, lo que implicaría que no puede ser resuelto por ningún algoritmo en tiempo polinomial (Sección 19.3.7 - Tim Roughgarden - Algorithms Illuminated). Además, se plantea que son los problemas que como mínimo tan difíciles como un problema $NP$, pero no necesariamente pueden pertenecer a este conjunto.

Definición formal: Un problema es $NP-Hard$ si todos los problemas en el conjunto $NP$ se puede reducir a este problema.

###¿Qué quiere decir que $P\neq NP$?

$P \neq NP$ corresponde a una conjetura, es decir, es algo que se cree que es verdad, pero no se ha probado. Recordemos que los problemas $P$ se refiere a los problemas fáciles de resolver con un algoritmo con tiempo polinomial. Mientras que los problemas $NP$ son los problemas que pueden ser verificados en tiempo polinomial si es que un resultado es correcto o no. Por lo tanto, si $P = NP $, quiere decir que **todos** los problemas $NP$ pueden ser resueltos en tiempo polinomial.


###¿Cómo se prueba  que un problema es NP-Hard usando **reducciones**?

La reducción es una técnica para demostrar que un problema inicial es tan difícil como otro, si podemos reducir el segundo problema al primero. Formalmente seria:

* Un problema $B$ que sea al menos tan difícil como un problema A $NP-Hard$ es un problema $NP-Hard$.

[Imagen extraida del libro: Tim Roughgarden - Algorithms Illuminated, sección: 19.5.1, tema: Reductions]

Al tener en claro la definición de la reducción, simplemente debemos encontrar un problema que sabemos de antemano y está comprobado que es $NP-Hard$ (problema A) e intentar **traducirlo** (Que una parte del algoritmo se adaptar al otro problema) a nuestro problema B, si podemos lograr esto, significa que nuestro problema original es como mínimo tan difícil como el otro problema que usamos para comprobar si nuestro problema es $NP-Hard$.

#2. Problemas NP-Hard escogidos

##2.1 Problema del Clique

###¿Qué es un Clique?:

Para hablar del problema mencionado, resulta necesario en primer lugar explicar en que consiste un **clique**.

Un **clique** o clan, se define como un grupo de $k$ nodos en un grafo indirecto $G = (V,E)$ que están conectados de tal manera a través de sus arcos que pueden formar un sub-grafo $G'\in G$ donde todos los nodos $V'\in V$ estén conectados entre sí por algún arco $E'\in E$.

El *tamaño* de un clique está representado por el número de nodos $k$ que pertenecen al sub-grafo.

###¿De qué trata el problema?:

El problema del clique es un problema de decisión, simplemente se pregunta al algoritmo si existe un clique de tamaño $k$ en el grafo. Este problema deriva de un problema de búsqueda que consiste en buscar un clique de tamaño máximo en un grafo, el cual se puede transformar en un problema de decisión si se le entrega la información al algortimo de que existe un clique de tamaño $k$ en el grafo.

Ejemplo de definición del problema:

$CLIQUE = $ \{ $(G,k) : G $ tiene un clique de tamaño $k$ \}

Vale destacar, que el *Problema del Clique* no sólo es un problema NP-Hard, sino también un NP-Complete. 

###Entradas y Salidas:

**Entrada**: Un grafo no dirigido $G$ y un índicador de tamaño entero $k$.

**Salida**: Un mensaje booleano (SI/NO) indicando si existe un clique de tamaño $k$ en el grafo $G$.

#2.2 Problema Vertex-Cover

##¿Que es el Vertex Cover?

Sea un grafo no dirigido $G = (V,E)$, se puede definir el Vertex-Cover como un subconjunto de nodos $V’ \subseteq V$, que debe seguir las siguientes restricciones:

* Tener un arco $(u,v) \in E$ 
* Que $u \in V’$ o $v  \in V’$ o que ambos al mismo tiempo pertenezcan al subconjunto de vértices.

En otras palabras, se define el Vertex Cover como el subconjunto de vértices o nodos que cubren o conectan con sus arcos incidente en $G$ de modo que abarquen todas las aristas en $E$, en síntesis, vértices que si o si debas transitar en el recorrido de cualquier grafo.

## Definición del problema
El problema que ocupa el Vertex Cover se enfoca en la optimización, puesto que se busca encontrar el Vertex Cover mínimo de tamaño k en un grafo.

$Entrada: $ un grafo $G = (V,E)$ no dirigido y un valor $k$.

$Salida: $ Un número $m$ que representa la cantidad de vértices que son Vertex Cover.

###Conjunto Independiente Máximo:

Sea de entrada un grafo $G = (V,E)$ y de salida un conjunto independiente con una cantidad máxima de nodos.

**Definición:**

- Dado un grafo G, un subconjunto de nodos $H ⊆ V(G)$ se le denomina conjunto independiente o estable si no hay dos vértices en H que sean adyacentes. Entonces el conjunto independiente máximo corresponde al mayor conjunto independiente que se puede obtener de un grafo G.

##3.1 Reducción del **Conjunto Independiente Máximo** al **Problema del Clique**.

- **Problema A**: Conjunto Independiente Máximo.
- **Problema B**: Problema del Clique. (Variante K-Clique).

###K-Clique:
Para generar la reducción del problema A, se debe realizar una modificación al problema del clique, en función de que este problema de decisión pase a entregar un resultado de búsqueda, como lo hace el Conjunto Independiente Máximo. Para hacer esto, se utilizará la variante del problema B, k-clique, que con los mismos parámetros de entrada que el algoritmo originalmente descrito, otorgará como salida una lista con los nodos pertenecientes al clique del grafo.

###¿Cómo y por qué es posible la reducción?

El problema A de esta reducción determina si existe una cantidad de nodos $|V|$ en un grafo no dirigido $G$ que no estén conectados entre sí, algo que parece completamente contrario a lo que haría un algoritmo que resolviera el problema B. Sin embargo, es esta particularidad la que permite generar un algoritmo que, usando B, pueda resolver a A.

Desarrollando la idea anterior, se puede buscar un clique de tamaño $k$ dentro de $G = (V,E)$, de modo que al estar los $k$ nodos de $V' \in V$ completamente conectados entre sí, se concluye que a lo más 1 de aquellos nodos podría pertenecer a la solución del problema A, ya que en el caso de considerar a otro nodo dentro del conjunto independiente, este ya no cumpliría con su condición principal.

Es por lo anterior, que es posible generar conjuntos de grupos interconectados con los cliques y el problema B aplicados de forma iterativa, y al momento de haber formado todos los grupos posibles, se podrá realizar un procedimiento extra para obtener la respuesta a A.

*Imagen Ilustrativa:*

Se debe continuar entonces, escogiendo un nodo por cada grupo, para añadir al conjunto independiente de nodos a formar para el problema A. Pero no se puede escoger cualquier nodo de $V'$, se debe asegurar en la selección, que los arcos de aquel nodo no conecten con algún nodo escogido anteriormente, ya que en aquel caso, se imcumpliría el requisito de independencia entre los nodos del conjunto.

Una vez se tiene un nodo correcto e independiente por cada clique, se habrá conseguido la respuesta al problema A, obteniendo la máxima cantidad de nodos no adyacentes entre sí, a raíz del problema B.

##3.2 Reducción del **Problema del Clique** al **Problema del Vertex Cover**.

Para demostrar que el Problema del Vertex Cover es $NP-Hard$ planteamos el siguiente teorema.

* $Teorema: $ El Vertex Cover es un problema $NP-Completo$.

Si planteamos que es $NP-Completo$, significa que el problema es tanto $NP$ como $NP-Hard$, por ende, se demostrará primero que es $NP$. 

* Teniendo un grafo $G = (V,E)$ no dirigido y un entero $k$.

El problema es $NP$, puesto que el certificado o posible solución $V’ \subseteq V$ es el Vertex Cover, para saber si corresponde o no a una opción válida se comprueba que $|V’| = k$ y finalmente se verifica en tiempo polinomial que cada arco $(u,v)  \in E$ que $u \in V’$ o $v \in V’$ o ambos al mismo tiempo pertenezcan a $V’$.

El problema es $NP-Hard$ porque podemos usar el problema de clique para demostrar que se puede reducir al nuestro, habiendo demostrado que Clique es un problema $NP-Completo$, nos daría a entender que es $NP$ y $NP-Hard$, en consecuencia, podemos llegar a establecer que el Vertex Cover también es $NP-Hard$.

La reducción de Clique a Vertex Cover se basa en el **complemento** que se obtiene a resolver al resolver Clique, dicho de manera más formal:

* Dado un grafo no dirigido $G = (V,E)$, se define al complemento de $G$ como $\bar{G}$ = $(V, \bar{E})$, donde $\bar{E}$ = {${(u.v) : u,v \in V, u \neq v, (u,v) \notin E}$.
* En resumen, el punto anterior quiere decir que $\bar{G}$ contiene las aristas que no están en $G$.

La reducción se aplica de la siguiente forma, teniendo el problema de Clique:

1.	Tomar la entrada $G$ y $k$.
2.	Calcular el complemento de $\bar{G}$ (se puede realizar en tiempo polinomial).
3.	Devolver $G$ y $|V|-k$, que corresponde al problema de Vertex Cover.


Para demostrar que el complemento de $G$ nos permite reducir de Clique a Vertex Cover planteamos lo siguiente:

* $Hipótesis:$ Supongamos que el grafo G tiene un Clique $V’\subseteq V$ con $|V'| = k$. Decimos que $V- V'$ es un Vertex Cover en G. 

Tenemos que $(u, v)$ es cualquier arco en $\bar{E}$ (cualquier arco que no esté en el grafo original o mejor dicho los no existentes). Entonces, $(u, v) \notin E$ en $G$, esto nos dice que al menos uno de los vértices no pertenece a $V'$, ya que todo par de vértices en $V'$ está conectado por una arista en el conjunto E. 

Imagen5.png

Paralelo a esto, al menos uno de $u$ o $v$ está en $V - V'$, lo que significa que el arco $(u, v)$ está en $V - V'$. Dado que el arco $(u, v)$  esta $\bar{E}$, cada arco en $\bar{E}$ está como vértice en $V - V'$. Por lo tanto, el conjunto $V - V'$ con tamaño |V|- k, forma un Vertex Cover en G, lo que indica que se puede reducir a partir Clique demostrando que el problema es $NP- Hard$. 

Finalmente, como se demostró que Vertex-Cover es $NP$ y $NP-Hard$, entonces el problema de Vertex-Cover es $NP-Completo$.

### Descripción del algoritmo que utiliza reducción para resolver el problema Clique usando Vertex-Cover

**Entrada de Clique:** La entrada del algoritmo para resolver el problema de Clique es un grafo no dirigido $G = (V,E)$ y un valor $k$ que indique la cantidad el tamaño del Clique a buscar.

**Preprocesamiento:** Como preprocesamiento no se debe hacer nada, dado que la entrada del Vertex-Cover es la misma que necesita el algoritmo de Clique.

**Procesamiento de Vertex-Cover:**

* 	**Entrada de Vertex-Cover:** un grafo $G = (V,E)$ no dirigido y un valor $k$ que indica el tamaño del Vertex.Cover.

*	**Procesamiento:** El algoritmo Vertex-Cover genera un conjunto $V’$ de nodos pertenecientes a $V$, tal que debe respetar la regla de $|V| -  k = |V’|$, dando que $|V’|$ es $m$.

*	**Salida de Vertex-Cover:** Un número $m$ indicando la cantidad de vértices que corresponden al Vertex-Cover mínimo, adicionalmente para resolver cuales son los vértices que corresponden a los Clique retornamos una lista $d$ con los nodos Vertex-Cover mínimos.

**Postprocesamiento:** Si valor obtenido de la siguiente operación: $|V| - m$ es mayor a cero retornamos un valor indicando ‘Si’ hay uno o más Clique en el grafo $G$, esa operación indica que si hay Vertex-Cover, entonces hay Cliques en $G$, si $m = |V|$ significa que no hay Cliques, debido a que todos los nodos son Vertex-Cover, por ende, se retorna ‘No’. Adicionalmente, si queremos saber que nodos son Clique, tomamos la lista auxiliar que retornamos del procesamiento de Vertex-Cover y mediante un ciclo extraemos los nodos pertenecientes a $d$ del grafo $G$, en $m$ iteraciones, obteniendo el grafo $G’$.

**Salida de Clique:** Un valor ‘Si / No’ indicando si existen Cliques en $G$, si es deseado, se puede retornar $G’$ que es $G$, pero sin los nodos Vertex-Cover.

# 4. Algoritmos propuestos para resolver los problemas escogidos

**Algoritmo k-Clique:**

Para determinar si un grafo de la forma $G = (V,E)$ contiene un clique de tamaño $k$, podemos listar todos los subconjutos de $V$ de tamaño $k$, e ir verificando cada subconjunto si se forma un clique de dicho tamaño.

Para poder implementarlo debemos crear una lista que guarde todas las secuencias de $k$. Una obtenidas dichas secuencias, iteramos en cada una para **verificar** si se considera un clique mediante combinatorias entre los vértices del grafo y el valor $k$. 


Tiempo de ejecución:
$O(k^2 * \binom{v}{k})$ ya que $k^2$ corresponde a la cantidad de arcos que hay que recorrer en cada subsecuencia y depende de la combinatoria entre los vértices y el tamaño de la subsecuencia.

**Algoritmo Vertex-cover:**

Una forma sencilla para resolverlo es implementando un algorito de "fuerza bruta".

El primer paso corresponde a crear una lista (llamemosla A) para guardar todas las combinaciones entre nodos y otra lista (llamemosla B) para almacenar los nodos en común entre combinaciones. Luego iteramos para obtener todas las combinaciones y la almacenamos en la lista. Después tomamos los nodos que se repiten en cada combinación y la almacenamos en su lista propia. Finalmente retornamos la cantidad de nodos que tiene la lista de nodos en común (B). Cabe señalar que se puede implementar un algoritmo para tener una "aproximación" de la cobertura de vértices del grafo en tiempo polinomial. Sin embargo, si queremos la cantidad exacta de nodos en común entra las combinaciones, el método de "fuerza bruta" descrito anteriormente es el más fiable.

Tiempo de ejecución: $O(2^v)$ siendo $v$ la cantidad de vértices. Esto es por la cantidad de iteraciones para obtener todas las combinaciones del grafo.
